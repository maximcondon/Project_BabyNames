#### BINS IS THE NUMBER OF SEPARATIONS IT HAS ie bin = 1 will be 1 block, 2 will be 2 blocks, 5 will be 5 blocks...

bins : integer or sequence, default 10
    
    Number of histogram bins to be used. If an integer is given, bins + 1 bin edges are calculated and returned. If bins is a sequence, gives bin edges, including left edge of first bin and right edge of last bin. In this case, bins is returned unmodified.
    
Histogram - To construct a histogram, the first step is to "bin" (or "bucket") the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval. The bins are usually specified as consecutive, non-overlapping intervals of a variable.

### Step 7

Try some optional parameters for the hist() function:

    facecolor='green',    ----- Changes colour to green
    facecolor='#ff0000',  ----- changed colour to red
    alpha=0.75,n          ----- fades colour on scale of 0 to 1 (0 being white)
    histtype='bar',       ----- already a bar I don't think it did anything

```python
lifeexp[2015].hist(bins=50, facecolor='#ff0000', alpha=0.75, histtype = 'bar')
plt.title('Histogram of World Life Expectancy in 2015')
plt.xlabel('Age (in years)')
plt.ylabel('Frequency')
plt.axis([45,85,0,15]) # First numbers are the x-axis, 2nd bunch of numbers are the y-axis
plt.show()
```

### Advantages and Disadvantages

**Pros of PNG**:
- minimum compression loss. The image quality is not changed by any compression ratio;
- format is suitable for storage of intermediate versions of the image. When you re-save image, quality is not lost;
- PNG supports a large number of colors. PNG-8 (256 colors) and PNG-24 (about 16.7 million. Colors);
- it supports multi level of transparency. Image has the 256 levels of opacity from fully opaque to fully transparent;
- it’s possible to work with layers;
- the ability to add to the file meta-data;
- small size files.

**Cons of PNG**:
- doesn’t support animation;
- ill-suited for working with full-color images;
- can not store multiple images in one file;

**Pros of SVG**:
- small size. SVG image elements take up much less space than their twins created in raster format;
- flexibility. With CSS, you can quickly change the graphics settings on the site, such as background color or the position of the logo on the page. To do this, you can edit the file in any text editor;
- scalability without changing the image quality. It is widely used for devices with screens Retina and those close to them;
- it’s possible to view the contents of the SVG file in any browser (IE, Chrome, Opera, FireFox, Safari, etc.).

**Cons of SVG**:
- the file size is growing very fast, if the object consists of a large number of small elements;
- it’s impossible to read a part of the graphic object, only the entire object and it slows you down.
