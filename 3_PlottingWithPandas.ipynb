|            Command             |           Description             |
|--------------------------------|:---------------------------------:|
|pandas.DataFrame.plot()	     | line plot of each column          |
|pandas.DataFrame.plot.bar()	 | one bar for each column           |
|pandas.DataFrame.plot.scatter() | creates a single scatter plot     |

```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline  

# matplotlib inline is a Jupyter Notebook-only command!
```

### What would be your first step?

We need data! We have two dimensions so we'll need some lists.

- We need list of data for x and y

```python
x = list(range(-10,11))
x
```

```python
y=[]

for i in x:
    y.append(i**2)
    
y
```

Alternatively, use List Comprehensions!!!

```python
y = [i**2 for i in x]
y
```

### Make sure that all variables have the same length

```python
if len(x) == len(y):
    print('Same length!')
else:
    print('Not matching!')
```

Or just use a boolean!

```python
len(x) == len(y)
```

Or can use assert to make sure that your code is correct and flag up errors!

```python
assert len(x) == len(y)
```

### Plot with pandas and matplotlib

There are more or less two ways to plot data:

1. Directly with matplotlib.pyplot (explicitly pass in X data, y data)

2. Through the Pandas DataFrame itself - df.plot() method!

```python
plt.figure(figsize = (8,5))
plt.plot(x,y)
plt.title('Reproduced Graph')
plt.savefig('polynomial.png')
```

#### Now let's do this just with Pandas and a DataFrame!

```python
pd.DataFrame(data=y, index=x)
```

### Regular plot (line/curve) with Pandas!

```python
df = pd.DataFrame(data=y, index=x)

df.plot()
```

### Bar Plot

```python
df.plot.bar()
```

### Horizontal Bar Plot

```python
df.plot.barh()
```

### Histogram!

```python
df.hist(bins=20)
```

### Boxplot!

```python
df.boxplot()
```
