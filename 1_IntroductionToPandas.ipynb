### Spiced Example

DataFrames are the central data structure in pandas. Each DataFrame consists of multiple columns, and a row and column index.

Columns of a DataFrame have the type `pd.Series`. Each Series contains only one type (in doubt, it assigns the generalized type object, which is usually not what you want).

Series is another word for column!

**e.g. from Kristian:**

```python
import pandas as pd

spices = ['One-Hot Chilli Peppers', 'Bayesian Basil', 'Tensor Thyme', 
          'Linear Lavendar', 'Aritificial Neural Nutmeg', 'Polynomial Peppermint']

participants = [2, 6, 9, 9, 9, 8]

df = pd.DataFrame({'name': spices, 'participants': participants})

print(df)
```

```python
print(df.describe())
```

```python
df.info()
```

```python
print(df.sort_values(by='participants'))
```

```python
print("\n total participants:", df ['participants'].sum())
```

### Boolean masks
- What do we do with Boolean masks? It returns a dataframe where the values returned satisfy the boolean condition.

```python
mask = countries['continent'] == 'Asia'
```

```python
countries[mask]
```

#### THIS IS AN EXCELLENT METHOD OF EXCLUDING POINTS FROM YOUR TABLE WITH A CONDITION!

Can be done in one line like this:

```python
countries [countries['continent'] == 'Asia']
```

### Can introduce filters:

```python
countries[countries['population'] > 200000000]
```

### Can do this for multiple conditions with a boolean & (and) and | (or)

```python
condition = (countries['population'] > 200000000) & (countries['continent'] == 'Asia')
```

```python
countries[condition]
```

### Slicing is a type of behaviour on collections!

Collections are lists, dictionaries, strings, pd.series, NumPy arrays (and sets and tuples but we'll see that later)

```python
peppermint = ['Eric', 'Josh', 'Karl', 'Max', 'Michael', 'Nedra', 'Anders', 'Daniela']
```

```python
len(peppermint)
```

```python
# first element of the list:
peppermint[0]
```

Slicing parameters: list[start_pos : stop_pos : step_size]

```python
peppermint[0:4]
```

```python
peppermint[0:4:2] #skips every step
```

Remember, strings are also lists, or collections!

```python
eric = 'Eric'
eric[0]
```

```python
eric[0:4:2]
```

#### We can slice in multi-dimensions!


```python
peppermint[0][0]
```

Here, we've gone into the list peppermint, gone to the first index position 0 and found 'Eric' and then we've taken the first index position 0 of 'Eric' which is 'E'!

### Create new column using a custom function
- (e.g. create a column that contains the first letter of the continent name)

```python
def first_letter(continent_name):
    return continent_name[0]
```

```python
countries['continent_first_letter'] = countries['continent'].apply(first_letter)
countries
```

```python
countries['continent'].apply(first_letter)
```

```python
countries['continent'][0]
```

### Advanced search functions using conditional statements

```python
countries['population'] #this gives me everything from that column
```

```python
countries[countries['population'] > 1000000000] #countries with pop. over 1 billion
```

#### Use an ampersand '&' to add extra conditions in the DataFrame!

```python
countries[(countries['population'] > 1000000000) & (countries['fertility'] > 2)] 
```

#### Use pipe '|' for OR condition in dataframe!

```python
countries[(countries.index == 'Bangladesh') | (countries.index == 'Mexico')]
```

### Iterate over rows and columns using iterative functions

#### For rows:

```python
count = 0

for i, value in countries.iterrows():
    if count < 3:
        print(i)
        print(value)
        count += 1
```

##### For columns - SUPER USEFUL FOR THE PROJECT

```python
for i, value in countries.iteritems():
    print(value)
```
